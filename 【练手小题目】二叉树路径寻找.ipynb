给出一个二叉树的节点列表Tree, 以及起始节点的值和终止节点的值
要求：创建二叉树，并找到起始节点到终止节点的路径

- 保证除了None以外，全部节点没有重复值
- 树的路径是唯一的

输入：
Tree = [3, 5, 1, 6, 2, 0, 8, None, None, 7, 4]
start = 5
end = 4

返回
[5, 2, 4]

解释说明：
- 

- 起点为5，终点为4的路径只有5->2->4


```python
from typing import List

class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
```

```python
class TreeFind:
    def __init__(self):
        self.result = []
        self.temp = []

    @staticmethod
    def bfs_tree(Tree: List):
        root = TreeNode(Tree.pop(0))
        temp = [root]
        while temp:
            for _ in range(len(temp)):
                node = temp.pop(0)
                if Tree and node:
                    node1 = TreeNode(Tree.pop(0))
                    node.left = node1
                    temp.append(node1)
                if Tree and node:
                    node1 = TreeNode(Tree.pop(0))
                    node.right = node1
                    temp.append(node1)
        return root

    def dfs(self, node: TreeNode, start, end):
        if self.temp and self.temp[-1].val == end:
            self.result = self.temp.copy()
        if node:
            if node.val == start or (self.temp and self.temp[0].val == start):
                self.temp.append(node)
            self.dfs(node.right, start, end)
            self.dfs(node.left, start, end)
            if self.temp:
                self.temp.pop()

    def FindTreePath(self, Tree: list, start: int, end: int):
        root = self.bfs_tree(Tree)
        self.dfs(root, start, end)
        return [x.val for x in self.result]

```

```python
Tree = [3, 5, 1, 6, 2, 0, 8, None, None, 7, 4]
start = 3
end = 4
tree_find = TreeFind()
tree_find.FindTreePath(Tree,start,end)
```
